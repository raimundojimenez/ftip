# Conclusion

Nous avons vu qu'en fonction de la caractéristique recherchée, les méthodes de la littérature scientifique sont très différentes.
* Pour les contours, il s'agit essentiellement d'un filtrage de l'image en utilisant le gradient ou le laplacien
  (filtres de Roberts ou Prewitt, détecteurs de Sobel ou Canny).
* Pour les coins, on effectue une mesure du changement d'intensité dans le voisinage des pixels (détecteur de Moravec et de Harris).
* Pour les lignes ou les cercles, l'image est transformée dans l'espace des paramètres de la forme géométrique (transformée de Hough).

Il existe d'autres approches adaptées à d'autres caractéristiques.
Par exemple, lorsqu'un motif, parfaitement connu sous forme de sous-image, est recherché,
alors l'intercorrélation de l'image avec ce motif est une technique très efficace
(elle est appelée « filtre adapté » ou en anglais : _matched filter_ ou [_template matching_](https://en.wikipedia.org/wiki/Template_matching)).

\begin{frame}{Filtre adapté}

La \emph{corrélation croisée} permet de détecter
un objet $g$ parfaitement connu (appelé \og{}motif\fg{})
dans une image $x$~:
\begin{equation*}
  R_{x,g}(u,v) = \sum_{m,n} x(m,n) g(u+m,v+n)
\end{equation*}
Elle peut s'écrire comme une convolution $\rightarrow$ \og{}Filtre adapté\fg{}.

Généralement, $x$ et $g$ sont remplacées par les images normalisées~:
\begin{equation*}
  x(m,n) \rightarrow \left( x(m,n)-\mu_x \right) / \sigma_x \qquad
  g(m,n) \rightarrow \left( g(m,n)-\mu_g \right) / \sigma_g
\end{equation*}
où $\mu_x$ et $\sigma_x$ sont respectivement la moyenne et l'écart-type de l'image $x$.
Cela aboutit à la \emph{corrélation croisée normalisée} qui est insensible aux changements d'amplitude.
