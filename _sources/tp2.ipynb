## Lossy compression

Le c\oe{}ur de la compression JPEG consiste à annuler les coefficients faibles
de la transformée en cosinus discrète (DCT) de l'image.
Dans cet exercice, une version simplifiée de la compression JPEG est mise en \oe{}uvre
en annulant les pixels de la DCT situés à l'extérieur d'un carré situé dans les basses fréquences et de côté~$C$.

\begin{questions}

  \item Calculez et affichez la DCT (\syntax{scipy.fftpack.dctn}) de l'image lena.tiff.
  Attention, utilisez l'option \syntax{norm='ortho'} de la fonction~!
  Pour simplifier la mise en \oe{}uvre, utilisez l'image en niveaux de gris.

  \item Le code ci-dessous génère un masque binaire de même taille que l'image~:
    \begin{code}
Mask = zeros(Y.shape)
Mask[:C,:C] = 1
    \end{code}
    Ce masque permet d'annuler les pixels à l'extérieur du carré~:
    affichez-le pour une valeur de $C$ choisie.

  \item Annulez les coefficients de la DCT à l'aide du masque généré.
