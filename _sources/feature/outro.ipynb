\begin{frame}{Filtre adapté}

La \emph{corrélation croisée} permet de détecter
un objet $g$ parfaitement connu (appelé \og{}motif\fg{})
dans une image $x$~:
\begin{equation*}
  R_{x,g}(u,v) = \sum_{m,n} x(m,n) g(u+m,v+n)
\end{equation*}
Elle peut s'écrire comme une convolution $\rightarrow$ \og{}Filtre adapté\fg{}.

\pause\medskip

Généralement, $x$ et $g$ sont remplacées par les images normalisées~:
\begin{equation*}
  x(m,n) \rightarrow \left( x(m,n)-\mu_x \right) / \sigma_x \qquad
  g(m,n) \rightarrow \left( g(m,n)-\mu_g \right) / \sigma_g
\end{equation*}
où $\mu_x$ et $\sigma_x$ sont respectivement la moyenne et l'écart-type de l'image $x$.
Cela aboutit à la \emph{corrélation croisée normalisée} qui est insensible aux changements d'amplitude.

\end{frame}

\begin{frame}{Filtre adapté}

L'inconvénient du filtrage adapté est qu'il est sensible aux variations d'orientation, de taille, etc.

\bigskip

On peut alors appliquer plusieurs filtres représentatifs de toutes les variations du motifs
$\Rightarrow$ très coûteux en temps de calcul~!

\bigskip

Alternative~: trouver des caractéristiques aux objets et effectuer une classification.

\end{frame}

\begin{frame}{Conclusion}

Détection de caractéristiques~: approches différentes en fonction de la caractéristique cherchée~!

\bigskip

\begin{itemize}
  \item Contour $\rightarrow$ filtrage de l'image en utilisant le gradient ou le laplacien (Roberts, Prewitt, Sobel, Canny~\dots)
  \item Coin $\rightarrow$ mesurer les changements d'intensité dans le voisinage des pixels (Moravec, Harris~\dots)
  \item Ligne, cercle $\rightarrow$ représenter l'image dans l'espace des paramètres (Hough~\dots)
  \item Motif bien connu $\rightarrow$ calculer une intercorrélation (filtre adapté)
\end{itemize}

\end{frame}
