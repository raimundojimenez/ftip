\begin{frame}{Nouvelle paramétrisation}

L'inconvénient de la paramétrisation $(a,b)$ est que l'espace des paramètres doit être borné et discrétisé
$\Rightarrow$ une droite verticale ($a=\infty$) ne peut pas être représentée.

\pause\medskip

$\Rightarrow$ Nouvelle paramétrisation

\begin{center}
  \includegraphics[page=1]{hough-new}
\end{center}

\vspace*{-2\baselineskip}

\begin{equation*}
  x = s \cos(\theta)
  \;\text{et}\;
  y = s \sin(\theta)
  \quad\Rightarrow\quad
  s = x \cos(\theta) + y \sin(\theta)
\end{equation*}

\end{frame}

% =============================================================================================== %

\begin{frame}{Nouvelle paramétrisation}

Pour chaque point $(x_i,y_i)$ de l'image, une sinusoïde est associée dans l'espace $(\theta,s)$~:
\begin{equation*}
  s = x_i \cos(\theta) + y_i \sin(\theta)
\end{equation*}

\begin{center}
  \includegraphics[page=2]{hough-new}
\end{center}

Les sinusoïdes correspondant aux points d'une même droite se coupent au point $(s^*,\theta^*)$ paramétrisant cette droite.

\end{frame}

\begin{frame}{Algorithme}

\begin{center}
  \setlength{\fboxsep}{3mm}
  \colorbox{algobg}{\parbox{.9\textwidth}{
  Appliquer une détection de contours \\
  Définir un accumulateur ($=$ espace des paramètres discrétisé) \\
  Pour chaque point des contours~: \\
  \albar Déterminer la sinusoïde correspondante au point \\
  \albar Incrémenter l'accumulateur le long de cette sinusoïde \\
  Rechercher les maxima de l'accumulateur \\
  En déduire les paramètres des lignes
}}
\end{center}

\end{frame}

\begin{frame}{Transformée de Hough}

Avantages~:
\begin{itemize}
  \item robuste au bruit
  \item robuste aux occlusions (peut détecter des objets partiellement recouverts)
  \item Extensible à tout objet paramétré (cercles, ellipses, \dots) \mycite{Duda72} \\
  \pause
  Exemple pour la détection de cercles~:
  \begin{equation*}
    (x-a)^2 + (y-b)^2 = r^2 \quad\rightarrow\quad \text{3 paramètres}
  \end{equation*}
\end{itemize}

\pause

Inconvénient~:
\begin{itemize}
  \item la dimension de l'accumulateur est égal aux nombres de paramètres
  $\Rightarrow$ le temps de calcul et la mémoire utilisée deviennent vite conséquents
\end{itemize}

\end{frame}
